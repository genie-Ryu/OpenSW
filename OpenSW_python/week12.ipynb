```python
# 실습 1) 서울의 일교차(2021.1.1 ~ 2021.12.31까지 자료 사용)
# 기온 공공 데이터에서 서울의 일교차가 가장 큰 날, 
# 최저 기온이 가장 낮은날, 최고 기온이 가장 높은 날을 찾아 출력

import csv

# 일교차가 가장 큰 날
def daily_tempRange(data):
    max_range = -999
    max_date = ''
    for row in data:
        highest_temp = float(row[-1])
        lowest_temp = float(row[-2])
        temp_range = highest_temp - lowest_temp
        if max_range < temp_range:
            max_range = temp_range
            max_date = row[0]
    return max_range, max_date

# 최저기온이 가장 낮은 날
def lowestDay(data):
    lowest_temp = 999
    lowest_date = ''
    for row in data:
        temp = float(row[-2])
        if lowest_temp > temp:
            lowest_temp = temp
            lowest_date = row[0]
    return lowest_temp, lowest_date

# 최고기온이 가장 높은 날
def highestDay(data):
    highest_temp = -999
    highest_date = ''
    for row in data:
        temp = float(row[-1])
        if highest_temp < temp:
            highest_temp = temp
            highest_date = row[0]
    return highest_temp, highest_date



# 1. 일교차가 가장 큰 날 출력
f = open('seoul_temperature.csv', 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')
header = next(data)
#print(header)

max_range, max_date = daily_tempRange(data)
print('일교차가 가장 큰 날\n{}\n'.format(max_date))

f.close()

# 2. 최저기온이 가장 낮은 날 출력
f = open('seoul_temperature.csv', 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')
header = next(data)

lowest_temp, lowest_date = lowestDay(data)
print('최저기온이 가장 낮은 날\n{}\n'.format(lowest_date))

f.close()

# 3. 최고기온이 가장 높은 날 출력
f = open('seoul_temperature.csv', 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')
header = next(data)

highest_temp, highest_date = highestDay(data)
print('최고기온이 가장 높은 날\n{}'.format(highest_date))

f.close()
```

```python
# 실습 2) 지하철 이용객(2021.12.1 ~ 2021.12.31까지 자료 사용)
# 노선별로 가장 이용객이 많은 3개 역과 가장 적은 3개 역 출력

import csv

# 이용객 수로 순위 정하기 함수
def ranking_station(subway_data):
    list2 = []
    for row in subway_data:
        row[-2] = commaToInt(row[-2])
        row[-1] = commaToInt(row[-1])
        passengers = row[-2] + row[-1]
        station = row[1] + ' ' + row[3]
        list1 = []
        list1.append(passengers)
        list1.append(station)
        list2.append(list1)
    list2.sort()
    return list2


# 콤마로 구분된 str 정수로 바꾸는 함수
def commaToInt(str):
    str = str.split(',')
    str2 = ''
    for i in str:
        str2 += i
    return int(str2)


# 결과 출력
file = open('subway.csv', 'r', encoding = 'cp949')
subway_data = csv.reader(file, delimiter = ',')
header = next(subway_data)

user_list = ranking_station(subway_data)

print('이용객이 가장 많은 역 순위')
for i in range(1,4):
    print(f'{i}. {user_list[-i][1]}({user_list[-i][0]})')

print('\n이용객이 가장 적은 역 순위')
for i in range(0,3):
    print(f'{i+1}. {user_list[i][1]}({user_list[i][0]})')
    
file.close()
```
